```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

from google.colab import files
uploaded = files.upload()

```

```python
img = cv2.imread('Crop_field_cropped.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray, 550, 690)
indices = np.where(edges != [0])
x = indices[1]
y = indices[0]

m = np.polyfit(x, y, 1)
fitted_line = np.poly1d(m)

fitted_y = fitted_line(x)
plt.scatter(x, y, c='blue', s=3, label='Original points')
plt.plot(x, fitted_y, c='red', label='Fitted line')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
